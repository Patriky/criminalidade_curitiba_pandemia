## Criando Dataframes com a taxa de crimes para cada Regional

Utilizando População e Regionais

```python
import pandas as pd 

df = pd.read_csv('../dados/dados_limpos/base_limpa_junho_2021.csv', sep=',', encoding="latin-1")
df['OCORRENCIA_DATA'] = pd.to_datetime(df['OCORRENCIA_DATA'])
df['OCORRENCIA_HORA'] = pd.to_datetime(df['OCORRENCIA_HORA'])
df['OCORRENCIA_DATA'] = df.OCORRENCIA_DATA.dt.normalize() # Para retirar o horário
df = df[df['OCORRENCIA_ANO'] >= 2020]
df = df[df['REGIONAL'] != 'REGIÃO METROPOLITANA']


# Abrindo o arquivo com as informações de população por regional
populacao = pd.read_csv("../dados/dados_limpos/renda_media_regionais.csv")

# Define o período de agrupamento
df['periodo'] = df.OCORRENCIA_DATA.dt.to_period("M")
df.sample(3)
```

```python
### Agrupando os tipos de crimes nos três tópicos  
roubo = df[(df["DESCRICAO1"] == "Roubo") | (df["DESCRICAO1"] == "Furto")]

violencia = df[(df["DESCRICAO1"] == "Violação de Medida Protetiva Lei Maria da Penha") | \
               ((df["DESCRICAO1"] != "Violação de Medida Protetiva Lei Maria da Penha") & (df["DESCRICAO2"] == "Violação de Medida Protetiva Lei Maria da Penha")) |\
               (df["DESCRICAO1"] == "Importunação\xa0sexual") |\
               (df["DESCRICAO1"] == "Estupro") |\
               (df["DESCRICAO1"] == "Exploração de menores") |\
               (df["DESCRICAO1"] == "Maus tratos à pessoas") |\
               (df["DESCRICAO1"] == "Abuso de incapazes")]

substancia = df[df["DESCRICAO1"] =="Substância Ilícita"]
```

# ROUBO 

```python
roubo_grouped = roubo.groupby(['periodo','REGIONAL'])['QTDE'].count().reset_index(name='ocorrências')
roubo_grouped.sample(3)
```

```python
# Defininco o período como index do tipo timestamp
roubo_grouped = roubo_grouped.set_index('periodo')
roubo_grouped.index = roubo_grouped.index.to_timestamp()
```

```python
# Criando dataframe para cada regional
roubo_bairro_novo = roubo_grouped[roubo_grouped['REGIONAL']=='BAIRRO NOVO']
roubo_boa_vista = roubo_grouped[roubo_grouped['REGIONAL']=='BOA VISTA']
roubo_boqueirao = roubo_grouped[roubo_grouped['REGIONAL']=='BOQUEIRÃO']
roubo_cajuru = roubo_grouped[roubo_grouped['REGIONAL']=='CAJURU']
roubo_cic = roubo_grouped[roubo_grouped['REGIONAL']=='CIC']
roubo_matriz = roubo_grouped[roubo_grouped['REGIONAL']=='MATRIZ']
roubo_pinheirinho = roubo_grouped[roubo_grouped['REGIONAL']=='PINHEIRINHO']
roubo_portao = roubo_grouped[roubo_grouped['REGIONAL']=='PORTÃO']
roubo_sta_felicidade = roubo_grouped[roubo_grouped['REGIONAL']=='SANTA FELICIDADE']
roubo_tatuquara = roubo_grouped[roubo_grouped['REGIONAL']=='TATUQUARA']

# Inserindo a população de cada regional. Não precisou dar merge com o df populacao
roubo_bairro_novo['populacao'] = 124330
roubo_boa_vista['populacao'] = 218741
roubo_boqueirao['populacao'] = 188192
roubo_cajuru['populacao'] = 202787
roubo_cic['populacao'] = 166192
roubo_matriz['populacao'] = 202304
roubo_pinheirinho['populacao'] = 121491
roubo_portao['populacao'] = 194675
roubo_sta_felicidade['populacao'] = 100762
roubo_tatuquara['populacao'] = 46149

# Fazendo a taxa de crimes por 100mil, para permitir comparação entre populações distintas
roubo_bairro_novo['taxa'] = roubo_bairro_novo['ocorrências']/ roubo_bairro_novo['populacao'] * 100000
roubo_boa_vista['taxa'] = roubo_boa_vista['ocorrências']/ roubo_boa_vista['populacao'] * 100000
roubo_boqueirao['taxa'] = roubo_boqueirao['ocorrências']/ roubo_boqueirao['populacao'] * 100000
roubo_cajuru['taxa'] = roubo_cajuru['ocorrências']/ roubo_cajuru['populacao'] * 100000
roubo_cic['taxa'] = roubo_cic['ocorrências']/ roubo_cic['populacao'] * 100000
roubo_matriz['taxa'] = roubo_matriz['ocorrências']/ roubo_matriz['populacao'] * 100000
roubo_pinheirinho['taxa'] = roubo_pinheirinho['ocorrências']/ roubo_pinheirinho['populacao'] * 100000
roubo_portao['taxa'] = roubo_portao['ocorrências']/ roubo_portao['populacao'] * 100000
roubo_sta_felicidade['taxa'] = roubo_sta_felicidade['ocorrências']/ roubo_sta_felicidade['populacao'] * 100000
roubo_tatuquara['taxa'] = roubo_tatuquara['ocorrências']/ roubo_tatuquara['populacao'] * 100000

# Deletando colunas desnecessárias e renomeando para o nome da regional
roubo_bairro_novo = roubo_bairro_novo.drop(['REGIONAL','ocorrências', 'populacao'], axis=1).rename(columns={'taxa': 'Bairro Novo'})
roubo_boa_vista = roubo_boa_vista.drop(['REGIONAL','ocorrências', 'populacao'], axis=1).rename(columns={'taxa': 'Boa Vista'}) 
roubo_boqueirao = roubo_boqueirao.drop(['REGIONAL','ocorrências', 'populacao'], axis=1).rename(columns={'taxa': 'Boqueirão'}) 
roubo_cajuru = roubo_cajuru.drop(['REGIONAL','ocorrências', 'populacao'], axis=1).rename(columns={'taxa': 'Cajuru'}) 
roubo_cic = roubo_cic.drop(['REGIONAL','ocorrências', 'populacao'], axis=1).rename(columns={'taxa': 'CIC'}) 
roubo_matriz = roubo_matriz.drop(['REGIONAL','ocorrências', 'populacao'], axis=1).rename(columns={'taxa': 'Matriz'}) 
roubo_pinheirinho = roubo_pinheirinho.drop(['REGIONAL','ocorrências', 'populacao'], axis=1).rename(columns={'taxa': 'Pinheirinho'}) 
roubo_portao = roubo_portao.drop(['REGIONAL','ocorrências', 'populacao'], axis=1).rename(columns={'taxa': 'Portão'}) 
roubo_sta_felicidade = roubo_sta_felicidade.drop(['REGIONAL','ocorrências', 'populacao'], axis=1).rename(columns={'taxa': 'Santa Felicidade'}) 
roubo_tatuquara = roubo_tatuquara.drop(['REGIONAL','ocorrências', 'populacao'], axis=1).rename(columns={'taxa': 'Tatuquara'}) 
```

```python
print(len(roubo_bairro_novo),len(roubo_boa_vista),len(roubo_boqueirao),len(roubo_cajuru),len(roubo_cic),len(roubo_matriz),len(roubo_pinheirinho),len(roubo_portao),len(roubo_sta_felicidade),len(roubo_tatuquara))
```

```python
# Agrupando os dataframes em um único dataframe
roubo = pd.merge(roubo_bairro_novo, roubo_boa_vista, how = 'left', left_index=True, right_index=True)
roubo = pd.merge(roubo, roubo_boqueirao, how = 'left', left_index=True, right_index=True)
roubo = pd.merge(roubo, roubo_cajuru, how = 'left', left_index=True, right_index=True)
roubo = pd.merge(roubo, roubo_cic, how = 'left', left_index=True, right_index=True)
roubo = pd.merge(roubo, roubo_matriz, how = 'left', left_index=True, right_index=True)
roubo = pd.merge(roubo, roubo_pinheirinho, how = 'left', left_index=True, right_index=True)
roubo = pd.merge(roubo, roubo_portao, how = 'left', left_index=True, right_index=True)
roubo = pd.merge(roubo, roubo_sta_felicidade, how = 'left', left_index=True, right_index=True)
roubo = pd.merge(roubo, roubo_tatuquara, how = 'left', left_index=True, right_index=True)
roubo['Tatuquara'] = roubo['Tatuquara'].fillna(0)
roubo
```
