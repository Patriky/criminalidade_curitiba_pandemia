```python
## Importação das libs e do dataframe
## Fonte dos dados: http://dadosabertos.c3sl.ufpr.br/curitiba/Sigesguarda/

import pandas as pd

df = pd.read_csv('../dados/2021-07-01sigesguarda.csv', sep=';', encoding="latin-1", low_memory=False)
df = df.drop(0) # Tirar a primeira linha, que é ------
df.head()
```

```python
df['ORIGEM_CHAMADO_DESCRICAO'].unique()
```

```python
#df[df.isna().any(axis=1)]
# Filtrando apenas as colunas mais relevantes
df = df[['ATENDIMENTO_ANO', 'ATENDIMENTO_BAIRRO_NOME', 'EQUIPAMENTO_URBANO_NOME', 'LOGRADOURO_NOME', 'NATUREZA1_DESCRICAO', 'NATUREZA2_DESCRICAO','SUBCATEGORIA1_DESCRICAO', 'OCORRENCIA_ANO', 'OCORRENCIA_MES','OCORRENCIA_DATA','OCORRENCIA_HORA', 'OCORRENCIA_DIA_SEMANA','ORIGEM_CHAMADO_DESCRICAO', 'REGIONAL_FATO_NOME']]
```

```python
df.columns   
```

```python
# Replace em dados nulos
df['ATENDIMENTO_ANO'] = df['ATENDIMENTO_ANO'].fillna(0)
df['ATENDIMENTO_BAIRRO_NOME'] = df['ATENDIMENTO_BAIRRO_NOME'].fillna("NÃO INFORMADO")
df['EQUIPAMENTO_URBANO_NOME'] = df['EQUIPAMENTO_URBANO_NOME'].fillna("NÃO INFORMADO")
df['LOGRADOURO_NOME'] = df['LOGRADOURO_NOME'].fillna("NÃO INFORMADO")
df['NATUREZA1_DESCRICAO'] = df['NATUREZA1_DESCRICAO'].fillna("NÃO INFORMADO")
df['NATUREZA2_DESCRICAO'] = df['NATUREZA2_DESCRICAO'].fillna("NÃO INFORMADO")
df['SUBCATEGORIA1_DESCRICAO'] = df['SUBCATEGORIA1_DESCRICAO'].fillna("NÃO INFORMADO")
df['OCORRENCIA_DIA_SEMANA'] = df['OCORRENCIA_DIA_SEMANA'].fillna("NÃO INFORMADO")
df['ORIGEM_CHAMADO_DESCRICAO'] = df['ORIGEM_CHAMADO_DESCRICAO'].fillna("NÃO INFORMADO")
df['REGIONAL_FATO_NOME'] = df['REGIONAL_FATO_NOME'].fillna("NÃO INFORMADO")

# Convertendo as colunas para o tipo certo
df['ATENDIMENTO_ANO'] = df['ATENDIMENTO_ANO'].astype(int)
df['OCORRENCIA_ANO'] = df['OCORRENCIA_ANO'].astype(int)
df['OCORRENCIA_MES'] = df['OCORRENCIA_MES'].astype(int)
df['OCORRENCIA_DATA'] = pd.to_datetime(df['OCORRENCIA_DATA'])
df['OCORRENCIA_HORA'] = pd.to_datetime(df['OCORRENCIA_HORA'])

df['QTDE'] = 1

df[df.isna().any(axis=1)]
#df = df.astype({'ATENDIMENTO_ANO': int, 'OCORRENCIA_ANO':int, 'OCORRENCIA_MES':int})
```

```python
#Renomear as colunas
df = df.rename(columns={'ATENDIMENTO_BAIRRO_NOME': 'BAIRRO', 'EQUIPAMENTO_URBANO_NOME': 'LOCAL_PUBLICO', 'LOGRADOURO_NOME': 'RUA',\
                       'NATUREZA1_DESCRICAO': 'DESCRICAO1', 'NATUREZA2_DESCRICAO':'DESCRICAO2',\
                        'SUBCATEGORIA1_DESCRICAO': 'SUBCATEGORIA', 'REGIONAL_FATO_NOME': 'REGIONAL'
                       })

df.info()
```
